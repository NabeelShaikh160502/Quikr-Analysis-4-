```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Bengaluru_House_Data.csv")
df
```

```python
df.shape
```

```python
df.info()
```

```python
for column in df.columns:
    print(df[column].value_counts())
    print("*"*50)
```

```python
df.drop(columns=['area_type','availability','society','balcony'])
```

```python
df.describe()
```

```python
df.info()
```

```python
df['location'].value_counts()
```

```python
df['location']=df['location'].fillna('Sarjapur  Road')
```

```python
df['location'].value_counts()
```

```python
df['size'].value_counts()
```

```python
df['size']=df['size'].fillna('2 BHK')
```

```python
df['bath']=df['bath'].fillna(df['bath'].median())
```

```python
df.info()
```

```python
df['bhk']=df['size'].str.split().str.get(0).astype('int')
```

```python
df[df.bhk > 20]
```

```python
df['total_sqft'].unique()
```

```python
def convert(x):
    temp=x.split('-')
    if len(temp)==2:
        return(float(temp[0])+float(temp[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df['total_sqft']=df['total_sqft'].apply(convert)
```

```python
df.head()
```

# Price per sqft(removing outliers)

```python
df['price_per_sqft']=df['price']*100000/df['total_sqft']
```

```python
df['price_per_sqft']
```

```python
df.describe()
```

```python
df['location'].value_counts()
```

```python
df['location']=df['location'].apply(lambda x:x.strip())
location_count=df['location'].value_counts()
```

```python
location_count
```

```python
location_count_less_10=location_count[location_count<=10]
location_count_less_10
```

```python
df['location']=df['location'].apply(lambda x:'other' if x in location_count_less_10 else x)
```

```python
df['location'].value_counts()
```

# Outlier handle

```python
df.describe()
```

```python
df=df[((df['total_sqft']/df['bhk'])>=300)]
df.describe()
```

```python
df.price_per_sqft.describe()
```

```python
upper_limit=df['price_per_sqft'].quantile(0.99)
upper_limit
```

```python
lower_limit=df['price_per_sqft'].quantile(0.01)
lower_limit
```

```python
new_df=df[(df['price_per_sqft']<=21707.94) & (df['price_per_sqft']>=2500.36)]
new_df
```

```python
new_df.describe()
```

```python
new_df.drop(columns=['area_type','availability','society','balcony'],inplace=True)
```

```python
new_df.shape
```

```python
new_df.drop(columns=['size','price_per_sqft'],inplace=True)
```

# Cleaned data

```python
new_df
```

```python
new_df.to_csv("Cleaned_data.csv")
```

```python
x=new_df.drop(columns=['price'])
y=new_df[['price']]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
column_trans=make_column_transformer((OneHotEncoder(sparse_output=False),['location']),
                                    remainder='passthrough')
```

```python
scaler=StandardScaler()
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lr=pipe.predict(x_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
#Decision tree
#random forest
#Lasso
#ridge
#ElasticÂ Net
```

# Apply lasso

```python
lasso=Lasso()
pipe=make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_lasso=pipe.predict(x_test)
r2_score(y_test,y_pred_lasso)
```

# Ridge

```python
ridge=Ridge()
pipe=make_pipeline(column_trans,scaler,lasso)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred_ridge=pipe.predict(x_test)
r2_score(y_test,y_pred_ridge)
```

# Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
decision=DecisionTreeRegressor()
pipe=make_pipeline(column_trans,scaler,decision)
pipe.fit(x_train,y_train)
```

```python
y_pred_decision=pipe.predict(x_test)
r2_score(y_test,y_pred_decision)
```

# RandomForest

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
randomforest=RandomForestRegressor()
pipe=make_pipeline(column_trans,scaler,randomforest)
pipe.fit(x_train,y_train)
```

```python
y_pred_randomforest=pipe.predict(x_test)
r2_score(y_test,y_pred_randomforest)
```

# ElasticÂ Net

```python
from sklearn.linear_model import ElasticNet
```

```python
elasticnet=ElasticNet()
pipe=make_pipeline(column_trans,scaler,elasticnet)
pipe.fit(x_train,y_train)
```

```python
y_pred_elasticnet=pipe.predict(x_test)
r2_score(y_test,y_pred_elasticnet)
```
